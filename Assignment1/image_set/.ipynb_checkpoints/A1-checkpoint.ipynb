### Part 1)

```python
red = np.zeros(img.shape)
blue = np.zeros(img.shape)
green = np.zeros(img.shape)
```

Seperate color channels

```python
for i in range (img.shape[1]):
    for j in range (img.shape[0]):
        if (i+j)%2 == 1:
            red[j,i] = img[j,i]
        elif j%2 == 0:
            blue[j,i] = img[j,i]
        else:
            green[j,i] = img[j,i]
```

Kernel Definition

```python
kernel_red = np.array([[0, 0.25, 0], [0.25, 1, 0.25], [0, 0.25, 0]])
kernel_bag = np.array([[0.25, 0.5, 0.25], [0.5, 1, 0.5], [0.25, 0.5, 0.25]])
```

```python
new_blue = cv2.filter2D(blue, -1, kernel_bag)
new_red = cv2.filter2D(red, -1, kernel_red)
new_green = cv2.filter2D(green, -1, kernel_bag)
```

Image Reconstruction

```python
new_img = cv2.merge(((new_blue), (new_green), (new_red)))
real_img.dtype
```

Calculating the difference

```python
dif = np.sqrt(np.square(real_img - new_img))
print(dif.dtype)
dif = dif.astype(np.uint8)
dif_sum_over_channels = cv2.cvtColor(dif, cv2.COLOR_BGR2GRAY)
```

Saving all together

```python
final_res_dif = np.concatenate([real_img, new_img, dif],axis = 1) 
```

Save the results

```python
cv2.imwrite('new_img.png', new_img)
cv2.imwrite('dif_img.png', dif)
cv2.imwrite('final_res_dif.png', final_res_dif)
```

### Part 2)

```python
green_red = new_green - new_red
blue_red = new_blue - new_red

green_red = green_red.astype(np.float32)
blue_red = blue_red.astype(np.float32)

median_green = cv2.medianBlur(green_red, 3)
median_blue = cv2.medianBlur(blue_red, 3)

second_green = ( median_green + new_red)
second_blue = ( median_blue + new_red)
```

```python
second_img = cv2.merge(((second_blue), (second_green), (new_red)))
```

```python
dif_r = np.sqrt(np.square(real_img[:,:,2] - new_red))
dif_g = np.sqrt(np.square(real_img[:,:,1] - second_green))
dif_b = np.sqrt(np.square(real_img[:,:,0] - second_blue))
second_dif = cv2.merge(((dif_b), (dif_g), (dif_r)))
```

```python
second_final_res_dif = np.concatenate([real_img, second_img, second_dif],axis = 1) 
```

```python
cv2.imwrite('second_img.png', second_img)
cv2.imwrite('second_dif_img.png', second_dif)
cv2.imwrite('second_final_res_dif.png', second_final_res_dif)
```

```python
second_dif = second_dif.astype(np.uint8)
dif_sum_over_channels = cv2.cvtColor(second_dif, cv2.COLOR_BGR2GRAY)
plt.imshow(dif_sum_over_channels)
```
